```python
w8 = pd.read_csv("OM.CompanionTables-W8.csv")
print(w8.shape)
w8.head()
```

```python
w8['PANGAEA sample identifier'] = w8['PANGAEA Sample ID']
```

```python
merged = w1[['Sample label [TARA_station#_environmental-feature_size-fraction]','Station identifier [TARA_station#]','Marine pelagic biomes (Longhurst 2007)','INSDC run accession number(s)','PANGAEA sample identifier','Size fraction lower threshold [micrometre]','Size fraction upper threshold [micrometre]']].merge(w8[['PANGAEA sample identifier','Mean_Lat*','Mean_Long*','Mean_Depth [m]*','Mean_Temperature [deg C]*','Mean_Oxygen [umol/kg]*','Mean Depth MLD Sigma [m]*']], on='PANGAEA sample identifier')
print(merged.shape)
merged.head()
```

```python
merged.tail()
```

```python
merged['depth_cat'] = [x.split('_')[2] for x in merged['Sample label [TARA_station#_environmental-feature_size-fraction]']]
```

```python
merged.head()
```

```python
merged.sort_values('Mean_Temperature [deg C]*')
```

```python
depths = merged[(merged['depth_cat']=='SRF')|(merged['depth_cat']=='MES')]
```

```python
depths.sort_values('Mean_Temperature [deg C]*')
```

```python
fracs = depths[depths['Size fraction lower threshold [micrometre]']=='0.22']
print(len(fracs))
fracs = fracs[fracs['Size fraction upper threshold [micrometre]']==3]
print(len(fracs))
```

```python
fracs.sort_values('Mean_Temperature [deg C]*')
```

```python
len(set(fracs['Station identifier [TARA_station#]']))
```

```python
stns = pd.DataFrame()
for stn in set(fracs['Station identifier [TARA_station#]']):
    subset = fracs[fracs['Station identifier [TARA_station#]']==stn]
    if len(subset)>=2:
        stns = stns.append(subset)
```
