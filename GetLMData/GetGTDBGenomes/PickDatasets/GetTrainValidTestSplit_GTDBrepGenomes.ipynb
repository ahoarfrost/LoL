```python
def count_children(node,count=0):
    for child in node.child_nodes():
        if child.taxon:
            count += 1
        count = count_children(child, count=count)
    return count
```

```python
def get_filenames(df):
    filenames = []
    for ix,row in df.iterrows():
        if row['ftp_path']:
            fname = str(row['ftp_path']).split('/')[-1]+'.fna'
            if fname=='nan.fna':
                fname = str(row['srr_accession'])+'.fna'
        filenames.append(fname) 
    return filenames
        
```

```python
def get_subset_leaves(node,to_match,taxa=[]):
    for child in node.child_nodes():
        if child.label:
            if len([s for s in to_match if s in child.label])>0:
                count=count_children(child,count=0)
                print('appending',count,'labels from node', child)
                these_taxa = [leaf.taxon.label for leaf in child.leaf_nodes()]
                taxa = taxa.extend(these_taxa)
        get_subset_leaves(child,to_match,taxa=taxa)
    return taxa
```

```python
def get_leaves(node,match,leaves=[]):
    for child in node.child_nodes():
        if child.label:
            if match in child.label:
                leaves = leaves.extend([leaf for leaf in child.leaf_nodes()])
        get_leaves(child,match,leaves=leaves)
    return leaves
```

```python
def print_children(node,match,max_count=0,min_count=0):
    for child in node.child_nodes():
        if child.label:
            if match in child.label:
                get_branch_children_counts(child,max_count,min_count)
        print_children(child,match,max_count,min_count)
        
def get_all_branch_children(node):
    for child in node.child_nodes():
        count=count_children(child,count=0)
        percent = count/23458.0
        print(child,'# children taxa:',count,'; level',child.level())
        get_all_branch_children(child)

def print_node_children_counts(node,match):
    for child in node.child_nodes():
        if child.label:
            if match in child.label:
                count=count_children(child,count=0)
                percent = count/23458.0
                print(child,'# children taxa:',count,'; level',child.level())
                get_all_branch_children(child)
        print_node_children_counts(child,match)
```

```python
import random 

def get_node(tree,match):
    for child in tree.seed_node.child_nodes():
        if child.label:
            if match in child.label:
                matched_node = child
        get_node(tree,match)
    return matched_node

def parse_genera_from_node(node,taxa,genomes_per_genera=1):
    for child in node.child_nodes():
        #if it's a genus, randomly select one genome from this genus to keep, delete the rest
        #(go delete route because there are occasionally unclassified taxa that don't have g__ labels)
        if child.label:
            if 'g__' in child.label:
                leaf_taxa = [leaf.taxon.label for leaf in child.leaf_nodes()]
                print(child.label,'# children taxa:',len(leaf_taxa))
                to_keep = random.sample(leaf_taxa,k=genomes_per_genera)
                to_delete = leaf_taxa
                [to_delete.remove(keeper) for keeper in to_keep]
                [taxa.remove(delete) for delete in to_delete]
                print('length new taxa list:',len(taxa))
        parse_genera_from_node(child,taxa,genomes_per_genera)
        

def parse_genera_from_tree(node,to_match,taxa,genomes_per_genera=1):
    for child in node.child_nodes():
        if child.label:
            if len([s for s in to_match if s in child.label])>0:
                print('processing',child.label,'...')
                parse_genera_from_node(child,taxa,genomes_per_genera=genomes_per_genera)
                print('length new taxa:',len(taxa))
        parse_genera_from_tree(child,to_match,taxa,genomes_per_genera=genomes_per_genera)
        
```

```python
def get_df_subset(ftp,arcftp,bac_list,arc_list):
    newdf_b = pd.DataFrame()
    for b in bac_list:
        if b[3:] in list(ftp['ncbi_genbank_assembly_accession']):
            newdf_b = newdf_b.append(ftp[ftp['ncbi_genbank_assembly_accession']==b[3:]])
        elif b[3:] in list(ftp['parsed_genome_accession']):
            newdf_b = newdf_b.append(ftp[ftp['parsed_genome_accession']==b[3:]])
        else:
            print('couldnt get record:',b)
            
    newdf_a = pd.DataFrame()
    for b in arc_list:
        if b[3:] in list(arcftp['ncbi_genbank_assembly_accession']):
            newdf_a = newdf_a.append(arcftp[arcftp['ncbi_genbank_assembly_accession']==b[3:]])
            #arcftp.drop(arcftp[arcftp['ncbi_genbank_assembly_accession']==b[3:]].index,inplace=True)
        elif b[3:] in list(arcftp['parsed_genome_accession']):
            newdf_a = newdf_a.append(arcftp[arcftp['parsed_genome_accession']==b[3:]])
            #arcftp.drop(arcftp[arcftp['parsed_genome_accession']==b[3:]].index, inplace=True)
        elif b in list(arcftp['parsed_genome_accession']): #for when only have SRR for e.g. UBA9915
            newdf_a = newdf_a.append(arcftp[arcftp['parsed_genome_accession']==b])
            #arcftp.drop(arcftp[arcftp['parsed_genome_accession']==b].index, inplace=True)
        else:
            print('couldnt get record',b)

    assert len(newdf_b)==len(bac_list)
    assert len(newdf_a)==len(arc_list)
    newdf = newdf_b.append(newdf_a,sort=True)
    return newdf
```
